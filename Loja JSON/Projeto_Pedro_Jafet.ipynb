```python
import json
import os.path
import sys
```

```python
def obter_dados():
    '''
    Essa função carrega os dados dos produtos e retorna uma lista de dicionários, onde cada dicionário representa um produto.
    NÃO MODIFIQUE essa função.
    '''
    with open(os.path.join(sys.path[0], 'dados.json'), 'r') as arq:
        dados = json.loads(arq.read())
    return dados

dados_json = obter_dados()

# Listar 10 itens para testar...
# dados_json[:10]
```

```python
def listar_categorias(dados):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    Essa função deverá retornar uma lista contendo todas as categorias dos diferentes produtos.
    Cuidado para não retornar categorias repetidas.    
    '''
    # Seu código
    categorias = []

    for dado in dados:
        if dado['categoria'] not in categorias:
            categorias.append(dado['categoria'])

    return categorias
```

```python
# Item 1 - Checagem da Listagem das Categorias
def selecao_item_1():
    categorias = sorted(listar_categorias(obter_dados()))
    lista_cat = []
    for cat in categorias:
        lista_cat.append(cat)
    return lista_cat
```

```python
def listar_por_categoria(dados, busca_categoria):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    O parâmetro "categoria" é uma string contendo o nome de uma categoria.
    Essa função deverá retornar uma lista contendo todos os produtos pertencentes à categoria dada.
    '''
    # Seu código
    
    produtos_da_categoria = []

    for dado in dados:
        if dado['categoria'] == busca_categoria:
            produto = dado['id']
            # print (f'Categoria {busca_categoria} ...... {produto}')
            produtos_da_categoria.append(produto) 

    return produtos_da_categoria


    
```

```python
# Item 2 - Checagem da Lista dos produtos em uma categoria

def selecao_item_2(categoria):

    dados = obter_dados()
    return (listar_por_categoria(dados,categoria))
```

```python
def produto_mais_caro(dados, busca_categoria):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    O parâmetro "categoria" é uma string contendo o nome de uma categoria.
    Essa função deverá retornar um dicionário representando o produto mais caro da categoria dada.
    '''
    # Seu código

    maior_preco = float(0)
    index = str()

    for dado in dados:
        if dado['categoria'] == busca_categoria:
            produto_preco = dado['preco']
            if float(produto_preco) > maior_preco:
                maior_preco = float(produto_preco)
                index = dado['id']
            #     print (f'--->{index} ... {busca_categoria} ... {produto_preco}')
            # else:
            #     print (f'Categoria {busca_categoria} ...... {produto_preco}')
    
    meu_dict = {'id': index, 'preco': maior_preco, 'categoria': busca_categoria}

    return meu_dict


```

```python
# Item 3 - Checagem do produto mais caro em uma categoria
def selecao_item_3(categoria):
    dados = obter_dados()
    return (produto_mais_caro(dados,categoria))
```

```python
def produto_mais_barato(dados, categoria):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    O parâmetro "categoria" é uma string contendo o nome de uma categoria.
    Essa função deverá retornar um dicionário representando o produto mais barato da categoria dada.
    '''
    # Seu código

    # Chama a outra função para saber qual o produto mais caro na categoria
    maior_preco_categoria = dict(produto_mais_caro(dados,categoria))
    # print ('=========',maior_preco_categoria)

    menor_preco = float(maior_preco_categoria['preco'])

    index = str()

    for dado in dados:
        if dado['categoria'] == categoria:
            produto_preco = dado['preco']
            if float(produto_preco) < menor_preco:
                menor_preco = float(produto_preco)
                index = dado['id']
            #     print (f'--->{index} ... {busca_categoria} ... {produto_preco}')
            # else:
            #     print (f'Categoria {busca_categoria} ...... {produto_preco}')
    
    meu_dict = {'id': index, 'preco': menor_preco, 'categoria': categoria}

    return meu_dict
    
```

```python
# Item 4 - Checagem do produto mais barato em uma categoria
def selecao_item_4(categoria):
    dados = obter_dados()
    return (produto_mais_barato(dados,categoria))
```

```python
def top_10_caros():
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    Essa função deverá retornar uma lista de dicionários representando os 10 produtos mais caros.
    '''
    # Seu código
    
    dados = obter_dados()
    dados_ordenados = sorted(dados, key = lambda item: float(item['preco']), reverse = True)
    top_10_caros = [dados_ordenados[i] for i in range(10)]

    return top_10_caros
```

```python
def top_10_baratos():
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    Essa função deverá retornar uma lista de dicionários representando os 10 produtos mais caros.
    '''
    # Seu código

    dados = obter_dados()
    dados_ordenados = sorted(dados, key = lambda item: float(item['preco']), reverse = False)
    top_10_baratos = [dados_ordenados[i] for i in range(10)]

    return top_10_baratos
```

```python
def menu():
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    Essa função deverá, em loop, realizar as seguintes ações:
    - Exibir as seguintes opções:
        1. Listar categorias
        2. Listar produtos de uma categoria
        3. Produto mais caro por categoria
        4. Produto mais barato por categoria
        5. Top 10 produtos mais caros
        6. Top 10 produtos mais baratos
        0. Sair
    - Ler a opção do usuário.
    - No caso de opção inválida, imprima uma mensagem de erro.
    - No caso das opções 2, 3 ou 4, pedir para o usuário digitar a categoria desejada.
    - Chamar a função adequada para tratar o pedido do usuário e salvar seu retorno.
    - Imprimir o retorno salvo. 
    O loop encerra quando a opção do usuário for 0.
    '''
    # Seu código
    # Menu principal

    selecao = ''

    while selecao != '0':
        selecao = input(('   SELECIONE UMA OPÇÃO\n1. Listar categorias\n2. Listar produtos de uma categoria\n3. Produto mais caro por categoria\n4. Produto mais barato por categoria\n5. Top 10 produtos mais caros\n6. Top 10 produtos mais baratos\n0. Sair'))
        print ('Opção escolhida:',selecao)
        if selecao == '1':
            print (selecao_item_1())
        elif selecao == '2':
            categoria = str(input('Digite uma categoria:'))
            if categoria != '':
                print ('Categoria:',categoria,'\n',selecao_item_2(categoria))
        elif selecao == '3':
            categoria = str(input('Digite uma categoria:'))
            if categoria != '':
                print (selecao_item_3(categoria))
        elif selecao == '4':
            categoria = str(input('Digite uma categoria:'))
            if categoria != '':
                print (selecao_item_4(categoria))
        elif selecao == '5':
            print (top_10_caros())
        elif selecao == '6':
            print (top_10_baratos())
        elif selecao != '0':
            print ('Opção incorreta, tente novamente!')
        
    return
```
